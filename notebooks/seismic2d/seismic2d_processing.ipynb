```python
acq_lines['cr286_line_length_km'] = acq_lines['cr286_line_length_km'].fillna(0)
print(acq_lines['cr286_line_id'].duplicated().any())
display(acq_lines.head(3))
cypher_seismic_line_txt = ""
for i in range(len(acq_lines)):
    _query = "MERGE (p:Seismic_2D_Line {{name: '{}', id: '{}', line_length: {}, survey: '{}'}})".format(
        acq_lines['cr286_line_name'][i], 
        acq_lines['cr286_line_id'][i], 
        acq_lines['cr286_line_length_km'][i], 
        acq_lines['cr286_seismic_2d_survey_id'][i], 
        )
    cypher_seismic_line_txt += _query + ';'
    if i < 3:
        print(_query)
```

#### 3. VESSELS

```python
vessels = vessels[vessels['cr286_vessel_id'].notna()]
vessels = vessels.reset_index(drop=True)
vessels.head(3)
vessels_id = vessels['cr286_vessel_id'].unique()
cypher_vessel_txt = ""
for i in range(len(vessels_id)):
    _query = "MERGE (p:Vessel {{name: '{}', id: '{}'}})".format(vessels_id[i], 'VID_'+str(i+1))
    cypher_vessel_txt += _query + ';'
    if i < 3:
        print(_query)
```

### 4. Relationships

```python
# Remove null values of cr286_block_id
blocks = blocks[blocks['cr286_block_id'].notna()]
blocks = blocks.reset_index(drop=True)
display(blocks.head(3))
cypher_bloc_survey_txt = ""
for i in range(len(blocks)):
    # WELL -> PROSPECT
    _query = """
    MATCH (s:Block {{id: '{}'}}), (o:Seismic_2D_Survey {{id: '{}'}})
    CREATE (s)-[:CÓ_KHẢO_SÁT {{record: '{}'}}]->(o)
    CREATE (o)-[:KHẢO_SÁT_TẠI {{record: '{}'}}]->(s)
    """.format(blocks['cr286_block_id'][i], 
                blocks['cr286_seismic_2d_survey_id'][i],
                blocks['cr286_record_id'][i],
                blocks['cr286_record_id'][i],
                )
    cypher_bloc_survey_txt += _query + ';'
print(_query)
```

```python
print(acq_lines['cr286_line_id'].duplicated().any())
display(acq_lines.head(3))
cypher_line_survey_txt = ""
for i in range(len(acq_lines)):
    # WELL -> PROSPECT
    _query = """
    MATCH (s:Seismic_2D_Line {{id: '{}'}}), (o:Seismic_2D_Survey {{id: '{}'}})
    CREATE (s)-[:THUỘC_KHẢO_SÁT]->(o)
    CREATE (o)-[:CÓ_TUYẾN]->(s)
    """.format(acq_lines['cr286_line_id'][i], 
                acq_lines['cr286_seismic_2d_survey_id'][i],
                )
    cypher_line_survey_txt += _query + ';'
print(_query)
```

```python
vessels = vessels[vessels['cr286_vessel_id'].notna()]
vessels = vessels.reset_index(drop=True)
display(vessels.head(3))
cypher_vessel_survey_txt = ""
for i in range(len(vessels)):
    # WELL -> PROSPECT
    _query = """
    MATCH (s:Vessel {{name: '{}'}}), (o:Seismic_2D_Survey {{id: '{}'}})
    CREATE (s)-[:KHẢO_SÁT]->(o)
    CREATE (o)-[:SỬ_DỤNG]->(s)
    """.format(vessels['cr286_vessel_id'][i], 
                vessels['cr286_seismic_2d_survey_id'][i],
                )
    cypher_vessel_survey_txt += _query + ';'
print(_query)
```
