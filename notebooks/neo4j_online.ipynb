```python
import os
from neo4j import GraphDatabase
from dotenv import load_dotenv; load_dotenv()

NEO4J_URI=os.environ['NEO4J_URI']
NEO4J_USERNAME=os.environ['NEO4J_USERNAME']
NEO4J_PASSWORD=os.environ['NEO4J_PASSWORD']
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))
```

```python
CYPHER_TEXT = '../data/Cypher.txt'
# Load and print single line of Cypher
with open(CYPHER_TEXT) as f:
    for line in f:
        print(line)
        driver.execute_query(query_=line, database='neo4j')
```

```python
import pandas as pd
RELATION_PATH = '../data/relations.csv'
relations_df = pd.read_csv(RELATION_PATH)
for i in range(len(relations_df)):
    # WELL -> PROSPECT
    w_p_query = """
    MATCH (s:Well {{name: '{}'}}), (o:Prospect {{id: '{}'}})
    CREATE (s)-[:KHOAN_VÀO_CẤU_TẠO]->(o)
    CREATE (o)-[:CÓ_GIẾNG_KHOAN]->(s)
    """.format(relations_df['WellAlias'][i], 
                relations_df['ProspectID'][i],
                )
    driver.execute_query(query_=w_p_query, database='neo4j')
    print(w_p_query)

    # WELL -> BLOCK
for i in range(len(relations_df)):
    w_b_query = """
    MATCH (s:Well {{name: '{}'}}), (o:Block {{id: '{}'}})
    CREATE (s)-[:THUỘC_LÔ]->(o)
    CREATE (o)-[:CÓ_GIẾNG_KHOAN]->(s)
    """.format(relations_df['WellAlias'][i], 
                relations_df['BlockID'][i],
                )
    driver.execute_query(query_=w_b_query, database='neo4j')
    print(w_b_query)
    
    # PROSPECT -> BLOCK
for i in range(len(relations_df)):
    p_b_query = """
    MATCH (s:Prospect {{id: '{}'}}), (o:Block {{id: '{}'}})
    CREATE (s)-[:THUỘC_LÔ]->(o)
    CREATE (o)-[:CÓ_CẤU_TẠO]->(s)
    """.format(relations_df['ProspectID'][i], 
                relations_df['BlockID'][i],
                )
    driver.execute_query(query_=p_b_query, database='neo4j')
    print(p_b_query)
```

```python
unique_values = """
MATCH (n:Investor)
RETURN DISTINCT n.name AS unique
"""
res = driver.execute_query(query_=unique_values, database='neo4j')
print(len(res[0]))
unique_v = []
for r in res[0]:
    unique_v.append(r['unique'])
print(unique_v)
```
